```
# Con los datos resultados del cluster del tipo:
data = [
('0.999', '23', '1'),
('0.2',   '20', '7'),
('0.1',   '15', '1'),
('0.399', '0',  '5'),
('0.499', '12', '4'),
('0.599', '25', '3'),
('0.799', '3',  '9'),
('0.899', '26', '5'),
]
data = sqlContext.createDataFrame(data, ['BI1', 'BI2', 'cluster'])

data = data.select(
data.BI1.cast('float'),
data.BI2.cast('float'),
data.cluster.cast('integer'))
data.show()
```

```
data_pd = data.toPandas()
ncols = len(data_pd.columns)
# data_pd.iloc[:,0:n-1]
data_BIDs = data_pd.iloc[:,0:2]
data_BIDs
```

```
# Aquí tenemos un dataframe para cada BID + la columna de cluster
import matplotlib.pyplot as plt

for bi in data_BIDs:
  df_BID =  data_pd[[bi, 'cluster']]

  plt.figure()
  bp = plt.boxplot(df_BID,widths = 0.6, by='cluster')

  for whisker in bp['whiskers']:
      whisker.set(color='#7570b3', linewidth=2)

  for flier in bp['fliers']:
      flier.set(marker='o', color='#e7298a', alpha=0.5)

  plt.ylim((0,250))
  plt.show()
  
```
